## Data Visualization

```python
df.groupby('year_added').size().plot(figsize=(12,5))
plt.title("Content Added to Netflix by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
plt.figure(figsize=(6,6))

df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
      startangle=90
)

plt.title("Movies vs TV Shows")
plt.ylabel("")
plt.show()
```

```python
df['rating'].value_counts().head(5).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Top 5 Content Ratings")
plt.ylabel("")
plt.show()
```

# Key Insights

1. Netflix contains more Movies (6131) than TV Shows (2676).

2. United States has the highest number of titles, followed by India and the United Kingdom.

3. TV-MA is the most common content rating.

4. Most content was released in 2018.

5. Netflix added the highest number of titles in 2019.

6. The most common genre is "Dramas, International Movies".

7. The dataset contains no duplicate records.

8. Director, Country and Cast columns have missing values.

9. Duration is stored in different formats:
   - Movies → Minutes
   - TV Shows → Seasons

## Conclusion

1. Netflix has significantly more Movies than TV Shows.
2. United States contributes the highest number of titles.
3. Most content was added in 2019.
4. Drama and International Movies are the most common genres.
